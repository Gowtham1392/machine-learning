```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Ice Cream Ratings.csv')
df = df.set_index('Date')
df
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings')
```

```python
df.plot(kind = 'bar', stacked  = True)
```

```python
df.plot.barh(stacked  = True)
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 100, color = 'Orange')
```

```python
df.plot.hist()
```

```python
df.boxplot()
```

```python
df.plot.area()
```

```python
df.plot.pie(y = 'Flavor Rating', figsize = (8,8))
```
